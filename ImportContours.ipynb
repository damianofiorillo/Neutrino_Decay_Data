#  Load modules

```python
import numpy as np
from matplotlib import pyplot as plt
import json
```

# Neutrino lifetime bounds from NGC 1068

## Importing the pre-computed 95% C.L. bounds

```python
# Using power-law flux
with open('digitized_contours/ngc1068_contours/PL_Prior_FixedFlavor.json', 'r') as json_file:
    ct_pl_prior_fixedflavor = json.load(json_file)
with open('digitized_contours/ngc1068_contours/PL_NoPrior_FixedFlavor.json', 'r') as json_file:
    ct_pl_noprior_fixedflavor = json.load(json_file)
with open('digitized_contours/ngc1068_contours/PL_Prior_FreeFlavor.json', 'r') as json_file:
    ct_pl_prior_freeflavor = json.load(json_file)
with open('digitized_contours/ngc1068_contours/PL_NoPrior_FreeFlavor.json', 'r') as json_file:
    ct_pl_noprior_freeflavor = json.load(json_file)

# Using power law with exponential cut-off
with open('digitized_contours/ngc1068_contours/PLCut_Prior_FixedFlavor.json', 'r') as json_file:
    ct_plcut_prior_fixedflavor = json.load(json_file)
with open('digitized_contours/ngc1068_contours/PLCut_NoPrior_FixedFlavor.json', 'r') as json_file:
    ct_plcut_noprior_fixedflavor = json.load(json_file)
with open('digitized_contours/ngc1068_contours/PLCut_Prior_FreeFlavor.json', 'r') as json_file:
    ct_plcut_prior_freeflavor = json.load(json_file)
```

## Neutrino flux: power law (PL)

### Prior on the total neutrino energy, fixed flavor composition

```python
#Converting to arrays 
ct_pl_prior_fixedflavor1=np.array(ct_pl_prior_fixedflavor['region1'])
ct_pl_prior_fixedflavor2=np.array(ct_pl_prior_fixedflavor['region2'])
ct_pl_prior_fixedflavor3=np.array(ct_pl_prior_fixedflavor['region3'])
#Plotting contours; each array contains the log_10(tau_2/m_2) and log_10(tau_3/m_3)
plt.plot(ct_pl_prior_fixedflavor1[:,0],ct_pl_prior_fixedflavor1[:,1],color='C0')
plt.plot(ct_pl_prior_fixedflavor2[:,0],ct_pl_prior_fixedflavor2[:,1],color='C0')
plt.plot(ct_pl_prior_fixedflavor3[:,0],ct_pl_prior_fixedflavor3[:,1],color='C0')
plt.xlabel(r'$\tau_2/m_2$ [s~eV$^{-1}$]')
plt.ylabel(r'$\tau_3/m_3$ [s~eV$^{-1}$]')
plt.xlim([min(ct_pl_prior_fixedflavor1[:,0]),4])
plt.ylim([min(ct_pl_prior_fixedflavor1[:,1]),4])
```

### No prior on the total neutrino energy, fixed flavor composition

```python
#Converting to arrays
ct_pl_noprior_fixedflavor1=np.array(ct_pl_noprior_fixedflavor['region1'])
ct_pl_noprior_fixedflavor2=np.array(ct_pl_noprior_fixedflavor['region2'])
#Plotting contours 
plt.plot(ct_pl_noprior_fixedflavor1[:,0],ct_pl_noprior_fixedflavor1[:,1],color='C0')
plt.plot(ct_pl_noprior_fixedflavor2[:,0],ct_pl_noprior_fixedflavor2[:,1],color='C0')
plt.xlabel(r'$\tau_2/m_2$ [s~eV$^{-1}$]')
plt.ylabel(r'$\tau_3/m_3$ [s~eV$^{-1}$]')
plt.xlim([min(ct_pl_prior_fixedflavor1[:,0]),4])
plt.ylim([min(ct_pl_prior_fixedflavor1[:,1]),4])
```
